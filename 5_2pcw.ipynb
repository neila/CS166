```
#coinflip

import numpy as np

p_head = 0.6
p_tail = 1-p_head
count_flip = 0
count_head = 0
count_tail = 0
position = [0]
n_trials = 3250

print(f"Initial position: {position[0]} \n")
while count_flip < n_trials:
  count_flip += 1
  if np.random.random() <= p_head:
    # print(f"flip {count_flip}: head")
    count_head += 1
    position.append(position[-1] + 1)
  else:
    # print(f"flip {count_flip}: tail")
    count_tail += 1
    position.append(position[-1] - 1)
  # print(f"position: {position[-1]} \n")  


print(f"{n_trials} flips. {count_head} heads, {count_tail} tails. \n position: {position[-1]} \n")  
```

```
#GENRALIZED MONTECARLO GAME

import random
experiments = 0
successes = 0
n_doors = 3
all_doors = [i+1 for i in range(n_doors)]
required_experiments = 50000  # Number of times to run the experiment

print_ = False

while experiments < required_experiments:
    if print_:
        print(f"EXPERIMENT {experiments+1}")

    # The car is hidden behind a random door.
    car_door = random.choice(all_doors)
    if print_:
        print(f"Car is in: {car_door}")

    # The contestant selects a random door.
    contestant_door = random.choice(all_doors)
    if print_:
        print(f"Contestant's initial choice: {contestant_door}.")

    # The host opens a different door that has a goat behind it.
    if contestant_door == car_door: #if contestant has already chosen car
        foo = all_doors.copy()
        foo.remove(contestant_door)
        keep = random.choice(foo) 
        foo.remove(keep)
        host_doors = foo
    else: #if contestant hasn't chosen car
        foo = all_doors.copy()
        foo.remove(contestant_door)
        foo.remove(car_door)
        host_doors = foo

    
    if print_:
        print(f"Host opened doors: {host_doors}.")

    # The contestant switches to the other door (not the one she chose originally and not the goat).
    no_contestant = all_doors.copy()
    no_contestant.remove(contestant_door)
    for i in host_doors:
      no_contestant.remove(i)
    contestant_new_door = no_contestant[0]
    if print_:
        print(f"Contestant switched to: {contestant_new_door}")

    # Did the contestant get the car? Count that as a success.
    if contestant_new_door == car_door:
        successes += 1
    experiments += 1
    if print_:
        print("\n")

print('The contestant won the car', successes / experiments, 'of the time')
```
